```python
import numpy as np
from scipy.signal import iirnotch, filtfilt, iirfilter

def filterNoise(all_signals, participant_ID):
    # Parameters
    fs = 1024  # Sampling frequency (Hz)
    f0 = 60.0  # Frequency to be removed (Hz)
    Q = 30.0  # Quality factor
    lowcut = 0.1  # Low cut-off frequency for bandpass filter (Hz)
    highcut = 30.0  # High cut-off frequency for bandpass filter (Hz)
    order = 4  # Filter order

    # Design the notch filter
    b_notch, a_notch = iirnotch(f0, Q, fs)
    
    # Design the bandpass filter
    b_bandpass, a_bandpass = iirfilter(order, [lowcut, highcut], rs=30, btype='band', analog=False, ftype='butter', fs=fs)

    # Initialize an empty array to hold the filtered data
    filtered_data = np.zeros_like(all_signals)

    # Apply the notch and bandpass filters to each electrode's data
    for i in range(all_signals.shape[0]):  # Loop through each electrode
        # Apply notch filter
        data_notch_filtered = filtfilt(b_notch, a_notch, all_signals[i, :])
        # Then apply bandpass filter
        filtered_data[i, :] = filtfilt(b_bandpass, a_bandpass, data_notch_filtered)

    # Save the array to a .npy file
    np.save('/oscar/data/brainstorm-ws/tfr_SerreLab_results/'+participant_ID+'_filtered_data.npy', filtered_data)

```

```python
import os
import re
import numpy as np

def preprocessEEG(base_directory, participant_id):
    # This pattern will match directories that start with a date and end with the participant ID
    pattern = re.compile(r"\d{4}-\d{2}-\d{2}_" + re.escape(participant_id) + r"_\d{2}")

    # Find all matching directories
    directories = [d for d in os.listdir(base_directory) if pattern.match(d)]

    # Initialize an empty list to store all signals
    all_signals = []

    for directory in directories:
        directory_path = os.path.join(base_directory, directory)
        file_list = os.listdir(directory_path)
        
        # Filter out unwanted files
        filtered_file_list = [f for f in file_list if f.endswith("pbz2") and not f.endswith("Events.pbz2") and f != ".DS_Store"]
        n_contact = len(filtered_file_list)
        timestep =checkLength(filtered_file_list,directory_path)
        
        # Initialize a numpy array to store signals from this directory
        signals = np.zeros((n_contact, timestep))  # Change the second dimension size if necessary

        for i, filename in enumerate(filtered_file_list):
            print(filename[-10:-5])
            
            filepath = os.path.join(directory_path, filename)
            # Load the EEG signal file; assuming loadsEEG is a function you've defined elsewhere
            signal = np.array(loadsEEG(filepath))
            signals[i] = signal

        # Add the signals from this directory to the overall list
        all_signals.append(signals)

    # Concatenate all the signals along the second axis
    if all_signals:
        combined_signals = np.concatenate(all_signals, axis=1)
    else:
        combined_signals = np.array([])  # Return an empty array if no signals were found

    return combined_signals
```
